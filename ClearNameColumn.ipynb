```python
import pandas as pd
```

### quoterchar delimits columns by the character ' (quotechar="'")

```python
df = pd.read_csv("https://www.openml.org/data/get_csv/20649148/freMTPL2freq.arff",quotechar="'")
```

### First let's understand how the dataframe is structured.

```python
df.head()
```

### the lambda function and replace are removing the " character from the column names

```python
df.rename(lambda x: x.replace('"',''), axis="columns", inplace=True)
df.head()
```

### head function displays the first 5 lines and shape displays the dataframe dimension

```python
display(df.head())
display("Dataset rows/cols: {}".format(df.shape))
```
